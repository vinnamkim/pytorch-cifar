```python
import torch
import os
import numpy as np
import matplotlib.pyplot as plt
from collections import OrderedDict
```

```python
datasets = {
    #'doublenorm' : torch.load('doublenorm_sharpness.stats'),
    #'norm2' : torch.load('norm2_sharpness.stats'),
    #'doublenorm2' : torch.load('doublenorm2_sharpness.stats'),
    'zerocenter' : torch.load('zerocenter_sharpness.stats'),
    'norm' : torch.load('norm_sharpness.stats')
}
```

```python
def plot(datasets, where, idx):
    for key in datasets:
        plt.plot(np.log(datasets[key][where][:, idx]), label=key)
    plt.title(where + str(idx))
    plt.legend()
    plt.show()

plot(datasets, 'train', 0)
plot(datasets, 'train', 1)
plot(datasets, 'valid', 0)
plot(datasets, 'valid', 1)
```
