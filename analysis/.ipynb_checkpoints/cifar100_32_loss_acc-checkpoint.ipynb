```python
import torch
import os
import matplotlib.pyplot as plt
import math
```

```python
batch_size = 32
```

```python
prefix_dir = '../cifar100_50_{0}'.format(batch_size)
```

```python
train_log_loss = {}
test_log_loss = {}
train_acc = {}
test_acc = {}

for d in sorted(os.listdir(prefix_dir)):
    m = d.replace('_50_{0}_10'.format(batch_size), '')
    if m == 'zerocenter2' or m == 'norm2':
        continue
    file_path = os.path.join(prefix_dir, d, 'stats.pth')
    stat = torch.load(file_path)
    train_log_loss[m] = [math.log(row[0]) for row in stat['train']]
    train_acc[m] = [row[1] for row in stat['train']]
    test_log_loss[m] = [math.log(row[0]) for row in stat['test']]
    test_acc[m] = [row[1] for row in stat['test']]
```

```python
colors = ['r', 'g', 'b', 'c', 'm', 'y']
```

```python
for c, m in zip(colors, train_log_loss):
    plt.plot(train_log_loss[m], color=c, label=m)
plt.title('train_log_loss')
plt.legend()
```

```python
for c, m in zip(colors, test_log_loss):
    plt.plot(test_log_loss[m], color=c, label=m)
plt.title('test_log_loss')
plt.legend()
```

```python
for c, m in zip(colors, test_acc):
    plt.plot(test_acc[m], color=c, label=m)
plt.title('test_acc')
plt.legend()
```

```python
for c, m in zip(colors, train_acc):
    plt.plot(train_acc[m][60:], color=c, label=m)
plt.title('train_acc')
plt.legend()
```

```python
for c, m in zip(colors, test_acc):
    plt.plot(test_acc[m][60:], color=c, label=m)
plt.title('test_acc')
plt.legend()
```
